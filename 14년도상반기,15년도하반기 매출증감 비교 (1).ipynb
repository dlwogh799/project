```python
import pandas as pd
import numpy as np
```

```python
season = pd.read_csv('./data/분기1.csv',encoding = 'cp949')
season
```

```python
season = season.fillna(0)
season = season.drop([season.index[0],season.index[1]])
season
```

```python
season = season.apply(pd.to_numeric)
season.dtypes
```

```python
season['14_1구매경험'] = np.where(season['구매횟수']>0,'유','무')
season['14_2구매경험'] = np.where(season['구매횟수.1']>0,'유','무')
season['14_3구매경험'] = np.where(season['구매횟수.2']>0,'유','무')
season['14_4구매경험'] = np.where(season['구매횟수.3']>0,'유','무')
season['15_1구매경험'] = np.where(season['구매횟수.4']>0,'유','무')
season['15_2구매경험'] = np.where(season['구매횟수.5']>0,'유','무')
season['15_3구매경험'] = np.where(season['구매횟수.6']>0,'유','무')
season['15_4구매경험'] = np.where(season['구매횟수.7']>0,'유','무')
season
```

```python
season = season.drop(['구매평균','구매평균.1','구매평균.2','구매평균.3','구매평균.4','구매평균.5','구매평균.6','구매평균.7'],axis=1)
season
```

```python
a = season[(season['14_1구매경험']=='유') & (season['14_2구매경험']=='유')
       & (season['14_3구매경험']=='유') & (season['14_4구매경험']=='유')
       & (season['15_1구매경험']=='유') & (season['15_2구매경험']=='유')
       & (season['15_3구매경험']=='유') & (season['15_4구매경험']=='유')]
a
```

14년도 상반기, 15년도 하반기 매출증감 비교

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
a14_12_avg = (a['총구매액']+a['총구매액.1']).sum()/18866
print(a14_12_avg,'\n')
a15_34_avg = (a['총구매액.6']+a['총구매액.7']).sum()/18866
print(a15_34_avg)

ratio = [8405737,8892839]
labels = ['14년상반기평균','15년하반기평균']
plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.show()
```

```python
a['14년상반기금액'] = a['총구매액']+a['총구매액.1']
a['15년하반기금액'] = a['총구매액.6']+a['총구매액.7']
a['매출증감'] = round((a['15년하반기금액'] - a['14년상반기금액']) / (a['15년하반기금액'] + a['14년상반기금액']) *100,1)
a['증감확인'] = np.where(a['매출증감']<2.8,'감소','증가')
grp=a.groupby(['증감확인'])
display(grp.describe())


ratio = [9640,9226]
labels = ['감소','증가']
plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.title('14년도상반기, 15년도하반기 증감비교',size=20)
plt.show()
print('감소: 9640(51.1%)','\n')
print('증가: 9226(48.9%)')
```
